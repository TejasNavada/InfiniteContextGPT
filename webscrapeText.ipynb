```python
import bs4
import requests

# function to scrape all paragraph text from website
def scrape_text_from_url(url):

    response = requests.get(url)

    # ensure good response
    if response.status_code == 200:
        # parse paragraphs
        soup = bs4.BeautifulSoup(response.text, 'html.parser')

        # get text from all <p> tags
        paragraphs = soup.find_all('p')
        # extract text from each paragraph and join it into a single string
        found_text = '\n'.join([paragraph.get_text() for paragraph in paragraphs])

        return found_text
    else:
        # return empty string upon a failed get request
        return ""
```

```python

# import required libraries
import numpy as np
from numpy.linalg import norm
import re
import torch
from sentence_transformers import SentenceTransformer
import json
import os
import random
import pytesseract
from pdf2image import convert_from_path
import glob



#get sentence transformer model
sentence_model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')


```

```python
openai_api_key = input("---------------------------------\nEnter OpenAI API KEY: ")

api_url = "https://api.openai.com/v1/chat/completions"

headers = {
    "Content-Type": "application/json",
    "Authorization": f"Bearer {openai_api_key}"
}






while True:
  url = input("---------------------------------\n(Type QUIT to exit)Enter Link: ")
  if( url =="QUIT"):
    break
  if(url == "PDF"):
    #get text from pdf
    fileName = input("---------------------------------\nEnter File Name: ")
    pdfs = glob.glob(fileName)

    url_text = ""

    for pdf_path in pdfs:

      pages = convert_from_path(pdf_path, 200)
      print("pages: ",len(pages))

      for pageNum,imgBlob in enumerate(pages):
        if(pageNum%10==0):
          print(pageNum)
        text = pytesseract.image_to_string(imgBlob,lang='eng')
        url_text+=text +".\n"
  else:
    #get text from webpage
    url_text = scrape_text_from_url(url)
    with open('url_text.txt', 'w') as fh:
      fh.write(url_text)
  if(url_text == ""):
    print("Link Failed")
    continue
  #split text into sentences
  sentences = re.split('\. |\.\n|\.\[', url_text)
  #need = ["I love eating pickled jalapenos for breakfast.  "]
  print(len(sentences)," sentences")
  print("loading...")
  #insert = int(random.random()*len(sentences))
  #sentences = sentences[0:insert]+need+sentences[insert:]

  #get sentence embeddings
  embeddings = sentence_model.encode(sentences)

  while True:
    question2 = input("---------------------------------\n(Type QUIT to exit)Enter Question: ")
    if question2 == "QUIT":
      break
      #get question embedding
    question_emb = sentence_model.encode(question2)
    #perform cosine similarity
    cosine = np.dot(embeddings,question_emb)/(norm(embeddings, axis=1)*norm(question_emb))
    outputs = cosine
    print("-----------------------------------------------")
    #get 10 most similar sentences and surrounding context
    sorted_idx = sorted(range(len(outputs)), key=lambda i: outputs[i], reverse=True)[:10]
    context = []
    for idx in sorted_idx:
      context = context + sentences[idx-2:idx+3]
    data = {
      "model": "gpt-3.5-turbo",
      "messages": [
          {
              "role": "system",
              "content": "You are a helpful assistant."
          },
          {
              "role": "user",
              "content": '\n'.join(context)+"\nGiven the context text above answer this question:"+question2
          }
      ]
    }
    #send to chatgpt and show response
    response = requests.post(api_url, headers=headers, json=data)
    # Check if the request was successful
    if response.status_code == 200:
        print(response.json()['choices'][0]['message']['content'])
    else:
        print("Error:", response.status_code, response.text)
```
